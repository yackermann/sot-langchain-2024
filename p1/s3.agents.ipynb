```python
from dotenv import load_dotenv
load_dotenv(dotenv_path='.env')
```

```python
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.chat_models import ChatOllama
from langchain_community.embeddings import OllamaEmbeddings 

# ----- OpenAI ----- #
chat_model = ChatOpenAI()
embeddings = OpenAIEmbeddings()

# ----- Ollama ----- #
# chat_model = ChatOllama()
# embeddings = OllamaEmbeddings()
```

```python
from langchain.agents import initialize_agent, AgentType, load_tools

from langchain.callbacks import StdOutCallbackHandler

tools = load_tools(["wikipedia"])

search_wiki_agent = initialize_agent(
    tools=tools,
    llm=chat_model,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
)

handler = StdOutCallbackHandler()

search_wiki_agent.run("Who is the PM of New Zealand?", callbacks=[handler])
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search = DuckDuckGoSearchRun()


tools = load_tools(["wikipedia"])
tools.append(search)

from langchain.globals import set_verbose, set_debug
set_debug(True)
set_verbose(True)


search_and_asnwer = initialize_agent(
    tools=tools,
    llm=chat_model,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
)

handler = StdOutCallbackHandler()

search_and_asnwer.run("Who is president of NZ? Ensure facts are clear", callbacks=[handler])
```

```python
from langchain.agents import tool

@tool
def current_time() -> str:
    """This function outputs the current time in the format: YYYY-MM-DD HH:MM:SS"""
    from datetime import datetime
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

tools = load_tools(["wikipedia"])
tools.append(current_time)


date_atent = initialize_agent(
    tools=tools,
    llm=chat_model,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
)

handler = StdOutCallbackHandler()

date_atent.run("What is the date today?", callbacks=[handler])
```
